```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
data = pd.read_csv('/kaggle/input/usahousing/USA_Housing.csv');
data.drop_duplicates(inplace=True)

```

```python
X = data[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
               'Avg. Area Number of Bedrooms', 'Area Population']]
y = data['Price']
```

```python
X = data.drop(['Price', 'Address'], axis=1)  # Assuming 'Address' is a feature and dropping it
y = data['Price']
```

random_state for splitting same data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
model=LinearRegression()
model.fit(X_train , y_train)
```

model training

```python
predictions = model.predict(X_test)
accuracy = model.score(X_test , y_test)
print("LR_Model acc :- ",accuracy*100)
```
